### Course Table

```python
query = "SELECT * FROM course"
course_data = pd.read_sql(query, conn)
```

```python
course_df = course_data.rename(columns={"CourseCode":"CourseCode__c",
                                             "CourseName":"CourseName__c",
                                             "CreditHours":"CreditHours__c",
                                             "BootCampCourse":"BootCampCourse__c",})
course_df.head()
```

```python
course_df = course_df[['CourseCode__c', 'CourseName__c', 'CreditHours__c', 'BootCampCourse__c']]
course_df.head()
```

```python
for index, row in course_df.iterrows():
    record = {
        'CourseCode__c': row['CourseCode__c'],
        'CourseName__c': row['CourseName__c'],
        'CreditHours__c': row['CreditHours__c'],
        'BootCampCourse__c': row['BootCampCourse__c']
    }
    try:
        sf.course__c.create(record)
    except Exception as e:
        print(e)
```

```python
course_lookup = []
data = sf.query_all_iter("SELECT CourseCode__c, Name FROM course__c")
for row in data:
    record = {'Course__c': row['Name'],
                'CourseCode__c': row['CourseCode__c']
             }
    course_lookup.append(record)
```

```python
course_lookup_df = pd.DataFrame(course_lookup)
course_lookup_df
```

### Class Table

```python
query = "SELECT c.*, co.CourseCode FROM class c INNER JOIN course co ON c.ID_Course = co.ID_Course"
class_data = pd.read_sql(query, conn)
class_data.head()
```

```python
class_df = class_data.rename(columns={"ID_Class":"ID_Class__c",
                                      "ID_Course":"ID_Course__c",
                                     "Section":"Section__c",
                                     "StartDate":"StartDate__c",
                                     "EndDate":"EndDate__c",
                                     "CourseCode":"CourseCode__c"}).drop(columns=['ID_Class__c', 'ID_Course__c'])
class_df.head()
```

```python
class_df = pd.merge(class_df, course_lookup_df, how='left')
class_df.head()
```

```python
class_df['StartDate__c'] = pd.to_datetime(class_df['StartDate__c']).dt.date
class_df['EndDate__c'] = pd.to_datetime(class_df['EndDate__c']).dt.date
class_df.head()
```

```python
for index, row in class_df.iterrows():
    record = {
        'Course__c': row['Course__c'],
        'Section__c': row['Section__c'],
        'StartDate__c':str(row['StartDate__c']),
        'EndDate__c':str(row['EndDate__c'])
    }
    try:
        sf.class__c.create(record)
    except Exception as e:
        print(e)
```

### Staff Assignment Table

```python
staffassignment_data['StartDate'] = pd.to_datetime(staffassignment_data['StartDate']).dt.date
staffassignment_data.head()
```

```python
staffassignment_df = staffassignment_data.rename(columns={"Role":"Role__c",
                                                            "StartDate":"StartDate__c",
                                                            "EndDate":"EndDate__c",
                                                            "Section":"Section__c",
                                                            "EmployeeID":"EmployeeID__c"}).drop(columns=['ID_StaffAssignment','ID_Staff', 'ID_Class'])
staffassignment_df.head()
```

```python
staffassignment_df = pd.merge(staffassignment_df, staff_lookup_df, how='left')
staffassignment_df.head()
```
